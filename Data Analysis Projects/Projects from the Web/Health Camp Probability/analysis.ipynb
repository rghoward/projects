```python
def calculate_attendance(category, Attended_1, Attended_2, Attended_3):
    a1 = 0
    a2 = 0
    a3 = 0    
    if category == 'First':
        if ~np.isnan(Attended_1):
            a1=1
    elif category == 'Second':
        if ~np.isnan(Attended_2):
            a2=1
    elif category == 'Third':
        if ~np.isnan(Attended_3):
            a3=1
    else:
        print('Category is wrong')
    return pd.Series([a1,a2, a3])      
```

```python
def RandomForest(X1,X2):
    from sklearn.ensemble import RandomForestClassifier as RFC
    x1 = X1.drop(columns=['Attended'])
    y1 = X1['Attended']
    rfc = RFC().fit(x1.drop(columns=['Patient_ID', 'Health_Camp_ID', 'First_Interaction']),y1)
    rfc_probs = rfc.predict_proba(X2.drop(columns=['Patient_ID', 'Health_Camp_ID', 'First_Interaction']))[:,1]
    X2['Attended'] = rfc.predict(X2.drop(columns=['Patient_ID', 'Health_Camp_ID', 'First_Interaction']))
    X2['Outcome'] = rfc_probs.tolist()
          
    return X2
```

```python
def Logistic_Regression(X1,X2):
    from sklearn.linear_model import LogisticRegression as LR
    x1 = X1.drop(columns=['Attended'])
    y1 = X1['Attended']
    lr = LR().fit(x1.drop(columns=['Patient_ID', 'Health_Camp_ID', 'First_Interaction']),y1)
    probs = lr.predict_proba(X2.drop(columns=['Patient_ID', 'Health_Camp_ID', 'First_Interaction']))[:,1]
    X2['Attended'] = lr.predict(X2.drop(columns=['Patient_ID', 'Health_Camp_ID', 'First_Interaction']))
    X2['Outcome'] = probs.tolist()
          
    return X2
```

```python
def Bayes(X1, X2):
    from sklearn.naive_bayes import GaussianNB
    x1 = X1.drop(columns=['Attended'])
    y1 = X1['Attended']
    bayes = GaussianNB().fit(x1.drop(columns=['Patient_ID', 'Health_Camp_ID', 'First_Interaction']),y1)
    probs = bayes.predict_proba(X2.drop(columns=['Patient_ID', 'Health_Camp_ID', 'First_Interaction']))[:,1]
    X2['Attended'] = bayes.predict(X2.drop(columns=['Patient_ID', 'Health_Camp_ID', 'First_Interaction']))
    X2['Outcome'] = probs.tolist()
          
    return X2
```

```python
def XGB(X1, X2):
    import xgboost as xgb
    x1 = X1.drop(columns=['Attended'])
    y1 = X1['Attended']
    dtrain = xgb.DMatrix(x1.drop(columns=['Patient_ID', 'Health_Camp_ID', 'First_Interaction']), label=y1)
    dtest = xgb.DMatrix(X2.drop(columns=['Patient_ID', 'Health_Camp_ID', 'First_Interaction']))
    param = {
    'max_depth': 10,  # the maximum depth of each tree
    'eta': 0.29,  # the training step for each iteration
    'silent': 1,  # logging mode - quiet
    'objective': 'multi:softprob',  # error evaluation for multiclass training
    'num_class': 2,
    'colsample_bytree': 0.5,
    'objective': 'reg:logistic'}  # the number of classes that exist in this datset
    num_round = 100  # the number of training iterations
    bst = xgb.train(param, dtrain, num_round)
    probs = bst.predict(dtest)
    probs = probs[:,1]
    X2['Outcome'] = probs.tolist()
    return X2
```

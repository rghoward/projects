```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
columns=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE',
         'DIS', 'RAD', 'TAX', 'PTRATIO', 'B-100(Bk-0.63)^2', 'LSTAT', 'MEDV']
house_df = pd.read_csv('boston.csv', names=columns)
house_df.head()
```

```python
from sklearn.linear_model import LinearRegression as LR
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
X = house_df.drop(columns='MEDV')
X = preprocessing.StandardScaler().fit(X).transform(X.astype(float))
y = house_df['MEDV']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
lr = LR().fit(X_train, y_train)
yhat = lr.predict(X_test)
```

```python
metrics.r2_score(y_test, yhat)
```

```python
np.sqrt(metrics.mean_squared_error(y_test, yhat))
```

```python
plt.scatter(yhat, y_test)
```

```python
yhat = rfr.predict(X_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor as KNR
```

```python
Ks = 10
scores = np.zeros(Ks-1)
for k in range(1,Ks):
    knr = KNR(n_neighbors=k).fit(X_train, y_train)
    yhat = knr.predict(X_test)
    score = metrics.r2_score(y_test,yhat)
    scores[k-1] = score
k = scores.argmax()+1
knr= KNR(n_neighbors=k).fit(X_train, y_train)
yhat=knr.predict(X_test)
metrics.r2_score(y_test, yhat)
```

```python
np.sqrt(metrics.mean_squared_error(y_test,yhat))
```

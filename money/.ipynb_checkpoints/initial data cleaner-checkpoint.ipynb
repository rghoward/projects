```python
def convertSplitFactor(x):
    if type(x) == float:
        return 0
    else:
        a, b = x.split(':')
        a=int(a)
        b=int(b)
        result = a/b
#         print (result)
        return result

```

```python
stocks.lastSplitFactor = stocks.lastSplitFactor.apply(convertSplitFactor)
```

```python
stocks[stocks.lastSplitFactor!=0].head()
```

```python
stocks.exDividendDate = stocks.exDividendDate.fillna(0)
stocks.head()
```

```python
def convertUTCtoDate(dividendDate, earningsDate):
    difference=0    
    if dividendDate != 0:
        difference = pd.to_datetime(earningsDate) - datetime.datetime.utcfromtimestamp(dividendDate) 
        difference = difference.days
    return difference
```

```python
stocks.head()
```

```python
i = 0
for row in stocks.itertuples():
    stocks.at[i, 'daysSinceLastDividend'] = convertUTCtoDate(row[7], row[13])
    
    i+=1
```
